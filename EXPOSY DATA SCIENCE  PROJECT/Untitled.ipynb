# Univariate Analysis

```python
# as the profit is a target variable we use the scatter plot.
plt.scatter(train.index,train['Profit'])
plt.show()
```

```python
sns.stripplot(y=train['Profit'])
```

```python
g = sns.pairplot(train, diag_kind="kde", markers="+",
                 plot_kws=dict(s=50, edgecolor="b", linewidth=1),
                 diag_kws=dict(shade=True))
```

```python
# correlation
train['Profit'].replace('no',0,inplace = True)
train['Profit'].replace('yes',1,inplace = True)
```

```python
corr = train.corr()
mask = np.array(corr)
mask[np.tril_indices_from(mask)] = False
fig, ax = plt.subplots()
fig.set_size_inches(20,10)
sn.heatmap(corr, mask =mask, vmax=.9, square = True,annot = True, cmap = "YlGnBu")
```

```python
# from the above graph we can infer that R&D spend is more.
```
